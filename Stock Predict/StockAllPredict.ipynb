## Section 1 - Basic Analysis of Stock Information

```python
#taking 3 different datasets for 3 different products 
dfap=df.loc[df['Product'] == 'APPL']
print(dfap.head())
dfsam=df.loc[df['Product'] == 'SAMSUNG']
print(dfsam.head())
dfmar=df.loc[df['Product'] == 'MARUTI']
print(dfmar.head())
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
dfap.plot('Date','Close',color='skyblue',figsize=(10,4))
dfsam.plot('Date','Close',color='red',figsize=(10,4))
dfmar.plot('Date','Close',color='green',figsize=(10,4))
```

Now that we've seen the DataFrame, let's go ahead and plot out the volume and closing price of the AAPL(Apple) stocks.

```python
# Let's see a historical view of the closing price of apple for instance
dfap['Close'].plot(legend='True',figsize=(12,6))
```

```python
# Now let's plot  
%matplotlib inline 
dfap.plot('Date','Volume',color='orange',figsize=(12,6))
```

### We can see that on Aug'2010 was the higher for APPL stock being traded.

Now that we've seen the visualizations for the closing price and the volume traded each day for appl stock. Let's go ahead and caculate the moving average for the AAPL stock.

For more info on the Moving Average(SMA & EMA) check out the following links:

1.) http://www.investopedia.com/terms/m/movingaverage.asp

2.) http://www.investopedia.com/articles/active-trading/052014/how-use-moving-average-buy-stocks.asp

```python
# Pandas has a built-in rolling mean calculator
# Let's go ahead and plot out several moving averages
MA_day = [10,20,50,100]
for ma in MA_day:
    column_name = 'MA for %s days' %(str(ma))
    dfap[column_name] = pd.rolling_mean(dfap['Close'],ma)
```

Now, lets plot all the additional Moving Averages for AAPL stock

```python
dfap[['Close','MA for 10 days','MA for 20 days','MA for 50 days','MA for 100 days']].plot(subplots=False,figsize=(12,6))
```

## Section 2 - Daily Return Analysis

Now, that we've done some baseline analysis, let's go ahead and dive a little deeper
Let's go ahead and use pandas to retrieve the daily returns for the APPL stock.

```python
# We'll use pct_change to find the percent change for each day
dfap['Daily Return'] = dfap['Close'].pct_change()

# Lets plot the daily return percentage
dfap['Daily Return'].plot(figsize=(10,4), legend=True, linestyle='--', marker='o')
```

```python
dfap.head()
```

Great, now let's get an overall look at the average daily return using a histogram. By using seaborn to create both a histogram and kde plot on the same figure.

```python
# only with histogram
dfap['Daily Return'].hist(bins=100)
```

```python
# Note the use of dropna() here, otherwise the NaN values can't be read by seaborn
sns.distplot(dfap['Daily Return'].dropna(), bins=100, color='magenta')
```

# Prediction starts

```python
from sklearn import preprocessing, cross_validation, svm
from sklearn.linear_model import LinearRegression
```

```python
import datetime as dt
dfap['Date'] = pd.to_datetime(dfap['Date'])
dfap['Date']=dfap['Date'].map(dt.datetime.toordinal)
dfap.head()
```

```python
X=(dfap.drop(['Product','Close','High','Low','Adj Close','Volume','Open'], axis=1))
# print(X)
X = np.array(X)
print(X)
# X = preprocessing.scale(X)
# print(X)
# df.dropna(inplace=True)
# df.head()
y = np.array(dfap['Close'])

X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.2,random_state = 0)
```

```python
# print(X_test)
```

```python
# dfap.head()
```

```python
clf = LinearRegression()
clf.fit(X_train, y_train)
predict = clf.predict(X_test)
print("\nPredicted values are...")
print(predict)
```

```python
confidence = clf.score(X_test, y_test)
print("\nTHE ACCURACY IS \n")
print(confidence)
```

```python
value = [737263]
vlaue = np.array(value)
print(value)
print(clf.predict(vlaue))
print(X_test[1])
print(clf.predict(X_test[1]))
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print(classification_report(y_train, clf.predict(X_train)))
print(classification_report(y_test, clf.predict(X_test)))
```

```python
accuracy_score(y_train, clf.predict(X_train))
accuracy_score(y_test, clf.predict(X_test))
```

```python
confusion_matrix(y_train, clf.predict(X_train))
```

```python
confusion_matrix(y_test, clf.predict(X_test))
```
